# TEST 4 - full grid using my composition

This notebook is only for studying the effects of Z and alpha

This is the same as in TEST 1 but for the whole grid. Now we are using the formula for the Y fraction.
In this test I include h2 and he3, use the tau100 tables and explore the following parameters:
- \[Fe/H\] = -0.5, 0, 0.3
- $\alpha_{MLT}$ = 1.0, 1.6, 1.82
- masses = (0.1, 0.6) in steps of 0.05

```python
import mesa_reader as mr
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D  
import matplotlib.cm as cm
from matplotlib.patches import Patch
```

```python
%matplotlib ipympl
```

```python
# Define the parameters
masses = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6]
alphas = [1.0, 1.6, 1.82]
metallicities = [0.0045, 0.0142, 0.028]


# Base path template
base_path = 'test2_grid_deuterium_tau100/results/mass{mass}_alpha{alpha}_z{z}/history.data'

# Create a list to store the MesaData instances and their associated alpha values
mesa_data_list = []

# Loop over the parameters to create the paths and MesaData instances
for mass in masses:
    for alpha in alphas:
        for z in metallicities:
            # Format the path
            path = base_path.format(mass=mass, alpha=alpha, z=z)
            
            try:
                # Create the MesaData instance
                mesa_data_instance = mr.MesaData(path)
                
                # Append the MesaData instance to the list
                mesa_data_list.append((mesa_data_instance, mass, z, alpha))
                
            except FileNotFoundError:
                print(f"FileNotFoundError: history.data not found for mass={mass}, alpha={alpha}, z={z}.")
```

```python
from scipy.interpolate import interp1d

# Create a list to store the MesaData instances and their associated parameters
mesa_data_list = []

# Loop over the parameters to create the paths and MesaData instances
for mass in masses:
    for alpha in alphas:
        for z in metallicities:
            # Format the path
            path = base_path.format(mass=mass, alpha=alpha, z=z)
            
            try:
                # Create the MesaData instance
                mesa_data_instance = mr.MesaData(path)
                
                # Append the MesaData instance and parameters to the list
                mesa_data_list.append((mesa_data_instance, mass, alpha, z))
                
            except FileNotFoundError:
                print(f"FileNotFoundError: history.data not found for mass={mass}, alpha={alpha}, z={z}.")

# Define the target age (in years)
target_age = 5 * 10**6

# Create dictionaries to store the Teff values at the target age for each combination of alpha and metallicity
teff_at_target_age = {(alpha, z): [] for alpha in alphas for z in metallicities}

# Iterate over MesaData instances in mesa_data_list
for mesa_data, mass, alpha, z in mesa_data_list:
    # Extract the age and Teff
    age = np.array(mesa_data.star_age)
    teff = np.array(mesa_data.Teff)

    # Check if the target age is within the age range of the data
    if target_age >= age[0] and target_age <= age[-1]:
        # Interpolate the Teff and radius values at the target age
        teff_interp = interp1d(age, teff, kind='linear')
        
        teff_at_age = teff_interp(target_age)
        
        # Append the interpolated Teff and radius values to the corresponding lists in the dictionaries
        teff_at_target_age[(alpha, z)].append(teff_at_age)
    else:
        print(f"Target age {target_age} years is out of range for mass={mass}, alpha={alpha}, z={z}.")

# Print the results for Teff at the target age
print("Teff values at target age:")
for (alpha, z), teff_list in teff_at_target_age.items():
    print(f"Alpha: {alpha}, Z: {z}")
    print(f"Teff values at {target_age} years: {teff_list}")
    print()
```

# D burning age 
Our objective is to plot D mass fraction vs age for all masses so we can state that the higher the mass the sooner the D is burned

```python
alpha = 1.82
z = 0.0142

# Create a dictionary to store the MesaData instances
mesa_data_dict = {}

# Loop over the masses to create the paths and MesaData instances
for mass in masses:
    # Format the path
    path = base_path.format(mass=mass, alpha=alpha, z=z)
    
    try:
        # Create the MesaData instance
        mesa_data_instance = mr.MesaData(path)
        
        # Store the instance in the dictionary
        mesa_data_dict[mass] = mesa_data_instance
        
    except FileNotFoundError:
        print(f"FileNotFoundError: history.data not found for mass={mass}, alpha={alpha}, z={z}")

# Print the results and create dynamic variables
for mass in masses:
    variable_name = f"mass{int(mass*100):03d}_z0142_alpha182"
    globals()[variable_name] = mesa_data_dict.get(mass)
    if globals()[variable_name] is not None:
        print(f"Accessed MesaData instance for {variable_name}")
    else:
        print(f"MesaData instance for {variable_name} not found")
```

```python
# Create a dictionary to store the ages when center_h2 drops below 2*10^-5
age_below_threshold = {}

# Iterate over the stored MesaData instances
for mass, mesa_data in mesa_data_dict.items():
    if mesa_data is not None:
        # Get the center_h2 and age arrays
        center_h2 = mesa_data.center_h2
        age = mesa_data.star_age
        
        # Find the first occurrence where center_h2 drops below 2*10^-5 (when the mass fraction drops below the initial value)
        below_threshold_indices = np.where(center_h2 < 2*10**(-5))[0]
        if below_threshold_indices.size > 0:
            first_below_threshold_index = below_threshold_indices[0]
            age_at_threshold = age[first_below_threshold_index]
            age_below_threshold[mass] = age_at_threshold
            print(f"Mass={mass}, Age when center_h2 < 2*10^-5: {age_at_threshold}")
        else:
            age_below_threshold[mass] = None
            print(f"Mass={mass}, center_h2 never drops below 10^-5")
    else:
        age_below_threshold[mass] = None
        print(f"Mass={mass}, no MesaData instance found")
```

```python
# Prepare data for plotting
masses_with_data = []
ages_with_data = []

for mass in masses:
    if age_below_threshold[mass] is not None:
        masses_with_data.append(mass)
        ages_with_data.append(age_below_threshold[mass])


# Plotting
plt.figure(figsize=(7, 5))

# plt.xscale('log')
plt.yscale('log')

plt.plot(masses_with_data, ages_with_data, marker='o', linestyle='-', color='b')

plt.xlabel(r'Mass ($M_\odot$)', fontsize=14)
plt.ylabel(r'D burning age (yr)', fontsize=14)

plt.xticks(fontsize=12) 
plt.yticks(fontsize=12)

#plt.title(r'Age vs. Mass for center H2 < 10$^{-5}$', fontsize=16)
plt.savefig('D-burning-age-vs-mass.png', dpi=300, bbox_inches='tight')
plt.show()
```
